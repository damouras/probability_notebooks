## Discrete Random Variables

Random Variables (RVs) represent quantities that depend on the outcome of a random experiment, and they offer a convenient way of working with probability models. Nevertheless, many realistic probability models are just too complex to handle analytically. In such cases, we rely on [computer simulations](https://en.wikipedia.org/wiki/Computer_simulation) of the random phenomenon in order to examine its behavior. But in order to simulate probability models, we must be able to [generate random numbers](https://en.wikipedia.org/wiki/Random_number_generation), i.e. instances or realizations of RVs. In real life, we can generate random numbers by flipping coins, rolling dice, [measuring cosmic radiation](https://phys.org/news/2004-09-random-cosmic-rays.html), or some other unpredictable process. In this notebook, we will look at how computers generate random numbers, and use them to confirm analytical results on discrete RVs. 


##### Pseudo-Random Numbers 

Computers are *deterministic* machines that cannot generate *truly* random numbers by running a program (although one can get better  numbers with appropriate [hardware](https://en.wikipedia.org/wiki/Hardware_random_number_generator)). But they can be programmed to efficiently generate sequences that, although deterministic, are highly irregular and unpredictable, i.e. you cannot figure out the next value by observing past values. The values thus generated are called [pseudo](https://en.wikipedia.org/wiki/Pseudo-)-random numbers, and the programs are called [pseudo-random number generators (PRNG)](https://en.wikipedia.org/wiki/Pseudorandom_number_generator). 

##### Linear Congruence Generators 

The most commonly used PRNGs until recently were [Linear Congruential Generators (LCG)](https://en.wikipedia.org/wiki/Linear_congruential_generator). The idea behind these is very simple: you pass a *linear recurrence* equation through a *modulo* operation:
$$ x_{n+1} = (a \cdot x_n + c ) \mod m ; \quad \quad n \ge 0$$
There are four parameters in a LCG: the *modulus* $m$, the *multiplier* $a$, the *increment* $c$, and the sequence's initial value $x_0$, called the *seed*. Below is an example for specific values $m=100, a=17, c=45, x_0=39$.

```python
m = 100; a = 17; c = 45; 
x = [0] * 25; x[0] = 39

for i in range( 1, len(x) ):
    x[i] = ( a * x[i-1] + c ) % m
print(x)
```

This generated a sequence of seemingly irregular numbers. If you look more closely though, you will see that the sequence repeats itself after revisiting the seed $x_0=x_{20}=39$! By construction, all LCGs will give repeating numbers at some point, and the length until they do that is called the *period*. You can show *every* LCG is periodic due to modular arithmetic: you can have at most $\# m $ distinct numbers from the $(\cdot) \mod m$ function, so by the [pigeonhole principle](https://en.wikipedia.org/wiki/Pigeonhole_principle) the seed will repeat after $\le m$ numbers and the sequence will reset. Note that the period of any LCG is at most $m$, and depending on the other parameters it can be much smaller (e.g., above we had $m=100$ but a period of $20$).
```python
from datetime import datetime

now = int( datetime.now().strftime("%S") )  

n=2**9
rnd_values = [ now % m ] * n

a = 13
c = 11
m = 2**7
for i in range(1,n):
    rnd_values[i] = ( a * rnd_values[i-1] + c) % m
    
plt.plot(rnd_values)
plt.vlines([2**7, 2**8, 2**7*3], 0, 128, linestyles ="dotted", colors ="k")
```

```python
plt.hist( rnd_values )
plt.show
```
